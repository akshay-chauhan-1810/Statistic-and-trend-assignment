```python
def plot_bar_population_by_continent():
    """
    plot the barplot for continent by population
    """
    by_continent = df[['Continent', '2022 Population']].sort_values(by='2022 Population', ascending=False)
   
    #plot bar plot
    plt.bar(by_continent['Continent'], by_continent['2022 Population'])
    plt.title('Population Timeline by Continent')
    plt.xlabel('Continent')
    plt.ylabel('Population')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
plot_bar_population_by_continent()
```

```python
def plot_box_Growth_by_Continent():
    """
    plot boxplot for World Population Growth by Continent
    """
    # Select relevant columns and make dataframe of them
    dataframe_for_box = df[['Continent', 'Growth Rate']]
    
    # Plotting
    plt.figure(figsize=(10, 6))

    # box plot
    dataframe_for_box.boxplot(by='Continent', column='Growth Rate', grid=False)
    plt.title('Growth Rate Distribution by Continent')
    plt.suptitle('')  # Remove default title generated by boxplot
    plt.xlabel('Continent')
    plt.ylabel('Growth Rate')
    plt.xticks(rotation=45)
    plt.show()
```

```python
plot_box_Growth_by_Continent()
```

```python
# select data frame time series of countries
Timeline_dataframe = pd.DataFrame({'Year':[1970,1980,1990,2000,2010,2015,2020,2022],
        'China_Population' : [822534450, 982372466, 1153704252, 1264099069, 1348191368, 1393715448, 1424929781, 1425887337],
        'United_States_Population':[200328340, 223140018, 248083732, 282398554, 311182845, 324607776, 335942003, 338289857],
        'Japan_Population':[105416839, 117624196, 123686321, 126803861, 128105431, 127250933, 125244761, 123951692],
        'India_Population' :[557501301, 696828385, 870452165, 1059633675, 1240613620, 1322866505, 1396387127, 1417173173],
        'United_Kingdom_Population':[55650166, 56326328, 57210442, 58850043, 62760039, 65224364, 67059474, 67508936]})
```

```python
def plot_line_of_sel_country():
    """
    plot the lineplot for population of selected countries over time
    """

    # plotting
    plt.figure(figsize=(10,6))

    #line plot
    plt.plot(Timeline_dataframe['Year'], Timeline_dataframe['India_Population'], label = 'India')
    plt.plot(Timeline_dataframe['Year'], Timeline_dataframe['Japan_Population'], label = 'Japan')
    plt.plot(Timeline_dataframe['Year'], Timeline_dataframe['United_States_Population'], label = 'United States')
    plt.plot(Timeline_dataframe['Year'], Timeline_dataframe['China_Population'], label = 'China')
    plt.plot(Timeline_dataframe['Year'], Timeline_dataframe['United_Kingdom_Population'], label = 'United Kingdom')
    
    plt.legend()
    plt.title("Population of Selected Countries over Time")
    plt.xlabel('Year')
    plt.ylabel('Population in Billions')
    plt.show()
```

```python
plot_line_of_sel_country()
```
